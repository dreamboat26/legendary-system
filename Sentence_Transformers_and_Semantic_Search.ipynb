**Generate Similarity Score With Sentence Transformers**

```python
from sentence_transformers import SentenceTransformer, util

# Load a pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example sentences
sentences = [
    "The sky is blue.",
    "I love to play soccer.",
    "Soccer is a popular sport worldwide.",
    "The ocean is vast and blue."
]

# Convert sentences to embeddings
embeddings = model.encode(sentences)

# Calculate similarity between sentences
similarity = util.pytorch_cos_sim(embeddings[0], embeddings[3])
print(f"Similarity between 'The sky is blue.' and 'The ocean is vast and blue.': {similarity.item():.2f}")
```

**Semantic Search Using Sentence Transformers**

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch
from datasets import load_dataset

# Step 1: Load the Quora Question Pairs dataset using Hugging Face
dataset = load_dataset("quora")

# Step 2: Convert it to a pandas DataFrame and extract relevant columns
data = pd.DataFrame(dataset['train'])

# Step 3: Extract question pairs from the 'questions' dictionary
data['question1'] = data['questions'].apply(lambda x: x['text'][0] if isinstance(x['text'], list) and len(x['text']) > 0 else None)
data['question2'] = data['questions'].apply(lambda x: x['text'][1] if isinstance(x['text'], list) and len(x['text']) > 1 else None)

# Step 4: Drop rows with missing data
data = data[['question1', 'question2']].dropna()

# Step 5: Prepare the list of documents
documents = data['question1'].unique().tolist()

# Step 6: Load the Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 7: Embed the documents
document_embeddings = model.encode(documents, convert_to_tensor=True, batch_size=64, show_progress_bar=True)

# Step 8: Define the semantic search function
def semantic_search(query, top_k=5):
    query_embedding = model.encode(query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, document_embeddings)[0]
    top_results = torch.topk(cosine_scores, k=top_k)

    print(f"\nQuery: {query}")
    print("\nTop results:")
    for score, idx in zip(top_results[0], top_results[1]):
        print(f"Score: {score:.4f} - Document: {documents[idx]}")

# Step 9: Test the semantic search
query = "How can I reset my password?"
semantic_search(query, top_k=5)
```
